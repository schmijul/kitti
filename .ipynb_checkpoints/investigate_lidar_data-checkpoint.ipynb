```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


```

```python

# Load the Lidar data as you have done before
idx = '0000000000'
bin_file = f'data/velodyne_points/data/{idx}.bin'
```

```python
lidar_numpy_data = np.fromfile(bin_file, dtype=np.float32).reshape(-1, 4)



# Extract x, y, z, and intensity
lidar_data_x = lidar_numpy_data[:,0]
lidar_data_y = lidar_numpy_data[:,1]
lidar_data_z = lidar_numpy_data[:,2]
lidar_data_intensity = lidar_numpy_data[:,3]
```

```python

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Use scatter for better performance with large point clouds
sc = ax.scatter(lidar_data_x, lidar_data_y, lidar_data_z, c=lidar_data_intensity, cmap='viridis', s=0.1)

# Set plot labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Lidar Point Cloud Visualization')

# Add color bar to visualize intensity
plt.colorbar(sc, label='Intensity')

# Show the plot
plt.show()
```
